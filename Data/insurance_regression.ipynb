```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the data
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# Extract independent varaiable, age, and dependent variables, charges
X = df[['age']].values  
Y = df['charges'].values  
```

```python
# Scatter plot of Age vs. Insurance Charges
plt.figure(figsize=(12, 6))
plt.scatter(X, Y, color='black',alpha=0.5)
plt.title('Scatter Plot of Age vs. Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

```python
# Perform linear regression
linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)

# Predicting insurance charges based on age
y_pred = linear_regressor.predict(X)
```

```python
# Plot the regression line with the data
plt.figure(figsize=(12, 6))
plt.scatter(X, Y, color='black', alpha=0.5, label='Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression: Age vs. Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.legend()
plt.grid(True)
plt.show()
```
