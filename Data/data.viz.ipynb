```python
# Import libraries
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('cars93.csv')
df.head()
```

```python
# Boxplot for revs per mile

plt.figure(figsize=(8, 6))
selected_makes = ["Audi", "Hyundai", "Toyota"]
subset_df = df[df["Manufacturer"].isin(selected_makes)]
sns.boxplot(x="Manufacturer", y="Rev.per.mile", data=subset_df)
plt.title("Revs per Mile for Selected Car Manufacturers")
plt.xlabel("Manufacturer")
plt.ylabel("Revs per Mile")
plt.show()

```

```python

# Check column names
print("Dataset columns:", df.columns)

# Verify required columns exist
if 'MPG.city' not in df.columns or 'MPG.highway' not in df.columns:
    raise KeyError("Columns 'MPG.city' or 'MPG.highway' not found in the dataset.")

# Histogram of MPG
plt.figure(figsize=(8,6))

# Plot City MPG
plt.hist(df['MPG.city'], bins=10, color='b', alpha=0.6, label='City')

# Plot Highway MPG
plt.hist(df['MPG.highway'], bins=10, color='r', alpha=0.6, label='Highway')

# Labels and Title
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('City vs Highway MPG')
plt.legend()
plt.show()
```

```python
# 3. Lineplot showing relationship between Wheelbase and Turning Circle
plt.figure(figsize=(8, 6))
sns.lineplot(x="Wheelbase", y="Turn.circle", data=df, marker="o")
plt.title("Relationship between Wheelbase and Turning Circle")
plt.xlabel("Wheelbase (inches)")
plt.ylabel("Turning Circle (feet)")
plt.show()
```

```python
# Barplot for mean horsepower
plt.figure(figsize=(8,6))

df_grouped = df.groupby("Type")["Horsepower"].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(df["Type"], df["Horsepower"])

# Labels and title
plt.xlabel("Car Type")
plt.ylabel("Average Horsepower")
plt.title("Average Horsepower by Car Type")

# Show plot
plt.show()
```

```python
"larger cars have more horsepower, but this may vary."
plt.figure(figsize=(8,6))
plt.scatter(df["Weight"], df["Horsepower"], alpha=0.6)
plt.xlabel("Weight (lbs)")
plt.ylabel("Horsepower")
plt.title("Relationship between Car Weight and Horsepower")
plt.show()
correlation = df[["Weight", "Horsepower"]].corr()
print(correlation)
```
